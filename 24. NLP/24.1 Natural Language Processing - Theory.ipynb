## How it's done ?

- We will want to :
    - Compile Documents in some fashion.
    - Featurize the Documents : Getting features from documents.
    - Compare the features.

## Simple Example :

**You have 2 documents:**
    - “Blue House”
    - “Red House”

A simple way to featurize text documents is to featurize based on word count. What we do is transform blue house into a vectorized word count. Where we create a vector count of all possible words through all the documents and count how many times those words are appearing in each document. 

- Featurize based on word count:
    - “Blue House” -> (red,blue,house) -> (0,1,1)
    - “Red House” -> (red,blue,house) -> (1,0,1)

## Mathematics Involved
A document represented as a vector of word counts is called a <span class="girk">**"Bag of Words"**</span>.
- “Blue House” -> (red,blue,house) --> (0,1,1)
- “Red House” -> (red,blue,house) --> (1,0,1)


- After we have bag of words vectors, we can use cosine similarity on vectors to determine similarity of documents themselves, as we are treating each document as a vector of features. 
- This enables us to perform mathematical operations such as cosine similarity where we take dot products and dividing that by product of magnitudes of vectors a and b.
- This is done to figure out how similar two text documents are to each other.
- We can improve on Bag of Words by adjusting word counts based on their frequency in corpus (the group of all documents)
- We can use **TF-IDF (Term Frequency - Inverse Document Frequency)** to do this.

## Terminology

- Term Frequency - Importance of the term within that document.
    - <span class="mark">TF(d,t) = Number of occurrences of term t in document d.</span>
- Inverse Document Frequency - Importance of the term in the corpus.
    - <span class="mark">IDF(t) = log(D/t)</span> where
        - D = total number of documents
        - t = number of documents with the term
