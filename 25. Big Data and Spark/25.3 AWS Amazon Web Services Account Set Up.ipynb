# AWS Amazon Web Services Account Set Up

## To create your AWS Account follow these steps :
1. Go to [AWS Free Tier Account](https://aws.amazon.com/free/)


2. Click on Create a Free Account.
3. Sign up with an email address.
4. Fill out profile information.
    - Contact Information
    - Billing Information
    - ID Verification
    - Choose free support plan


- AMAZON EC2 is basically a less powerful cloud computer and we will setup anaconda, scala, jre everything there.
- The instance of EC2 that we will be using ini tc2.micro. Make sure you select this to avoid getting charged on the debit/credit card you entered for verification.
- Also, EC2 instance in free tier only allows 750 hours per month so always terminate your instance after you are done with your work.
- AWS has other products like S3, SageMaker etc. We may use larger instance of a computer when actually working for an organization, for now we might as well stay within the limits using tc2.micro of EC2.
- Setting up a larger instance would be same as setting up micro, which we are doing here.
