## Local vs Distributed 


### Local Machine : 
- This is what we have been using, it is our own computer or machine where we have restricted RAM, processing power, GPU power etc.

### Distributed System :
- We can have one machine controlling a distribution of multiple machines.

Following image illustrates how having local vs distributed machine can help to harness power of multiple machines.
On one local machine we might use all its cores, however on a distributed machine we can have one smaller master node, controlling a distributed system of computers.

- This allows us to take advantage of distributed system and retain more processing power and memory then we would have done on local machine. We discuss more about this as we go towards Hadoop and MapReduce.


# Big Data
- A local process uses the computation resource of a single machine.
- A distributed process has access to the computational resources across a number of machines connected through a network.
- After a certain point, it is easier to scale out to many lower CPU machines, than it is to try to scale up to a single machine with a high CPU.
- Distributed machines also have the advantage of easily scaling, as we can just add more machines.

- These also include fault tolerance,redundancy, as failure of one machine doesn't affects the network, and whole network can still go on.

- Let's have a look on typical format of a distributed architecture which uses Hadoop.

## Hadoop

- Hadoop is a way to distribute very large files across multiple machines.
- It uses the Hadoop Distributed File System (HDFS)
- HDFS allows a user to work with large datasets.
- HDFS also duplicates blocks of data for fault tolerance
- It also then uses MapReduce
- MapReduce allows computations on that data

## Distributed Storage - HDFS



- There is main (name) node with CPU and RAM attached to it.
- There are several other distributed machines known as data nodes.

______
- HDFS will use blocks of data, with a size of 128 MB by default.
- Each of these blocks are replicated 3 times.
- The blocks are distributed in a way to support fault tolerance i.e. if one of the lower data nodes goes corrupt for some reason, then also data is replicated on other nodes, leading to no loss of information.

- Smaller blocks provide more parallelization during processing.
- Multiple copies of a block prevent loss of data due to failure of a node.

## MapReduce

- MapReduce is a way of splitting a computation task to a distributed set of files (such as HDFS).
- It consists of a Job Tracker and multiple Task Trackers.

- The Job Tracker sends the code to run on the Task Trackers.
- The Task Trackers allocate CPU and memory for the tasks and monitor the tasks on the worker nodes.

## Wrapping Up:
- What we covered can be thought of in two distinct parts :
    - Using HDFS to distribute large data sets.
    - Using MapReduce to distribute a computational task to a distributed data set.
