# Random Forests

As a way to improve performance of single decision trees.


- The primary weakness of decision trees is that they do not tend to have best predictive accuracy.
- This is due to the high variance, that is different splits in training data, leads to very different trees.
- Bagging is a general purpose procedure to reduce the variance of machine learning method. More on this in Chapter 8 of Reading assignment.
- We can build from the idea of bagging by utilising Random Forests.
- Random forest are just a small variation of these bagged tress and have better performance.
- We create an ensemble of decision trees using bootstraped samples of training sets.
- Bootstrap sample of training set means sampling from training set with replacement.
- When building each tree, each time a split is considered a random sample of m features is chosen as a split candidate, from a full set of p features. Split only allows to use one of the m features.
- A new random sample of features is chosen for every single tree at every single split
- For classification, m is typically chosen to be the square root of p.

## What's the point of using random forests ?
- Say there is one **very strong feature** in the dataset. A feature which is very strong in predicting a certain class.
- When using **"bagged trees"**, that is the bootstrap sampling, most of the trees will use that feature as the top split, resulting in an ensemble of similar trees that are highly correlated.
- Averaging highly correlated quantities does not significantly reduce variance.
- By randomly leaving out candidate features from each split Random Forest **"decorrelates"** the trees(making the trees independent of each other), such that the averaging process can reduce the variance of the resulting model.
- So that we won't be much affected by the features that much affect the class data.
