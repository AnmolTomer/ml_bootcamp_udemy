## About K Means Clustering Algorithm :

- Allows us to cluster unlabelled data in an unsupervised machine learning algorithm.  
- K Means Clustering is an unsupervised learning algorithm that will attempt to group similar clusters together in your data. 

## So what does a typical clustering problem look like? 

* Cluster similar documents.
* Cluster Customer based on features
* Market Segmentation
* Identify similar physical groups


* The overall goal is to divide the data into distinct groups such that observations within each group are similar.
* Above image illustrates K-means clustering trying to cluster the data into 5 different colour clusters.

## K-Means Algorithm : How does it actually works ?

* Choose a number of Clusters K.
* Randomly assign each point to a cluster.
* Until clusters stop changing, repeat the following:
    * For each cluster, compute the cluster centroid by taking the mean vector of points in the cluster.
    * Assign each data point to the cluster for which the centroid is the closest.

## Choosing a K value

* As discussed above, to perform K means clustering we have to decide how many clusters we expect in the data.



* There is no easy answer for choosing a best K value.
* One way is the elbow method.

* First of all, compute the sum of squared error (SSE) for some values of k (for example 2, 4, 6, 8, etc.)
* The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid. 
* If you plot k against the SSE, you will see that the error decreases as k gets larger; this is because when the number of clusters increases, they should be smaller, so distortion is also smaller.
* The idea of the elbow method is to choose the k at which the SSE decreases abruptly. 
* This produces an "elbow effect" in the graph, as you can see in the following picture:

- The above image depicts plotted out elbow method, on the horizontal (X-axis) we have the K value. Which stands for number of clusters we choose. For the algorithm.
- On y-axis we have within groups sum of Squares. 
- We want to have that k value after which even if we increase number of clusters k there won't be much of an information gain. Meaning we wouldn't be significantly decreasing within groups SSE by increasing K. 
- Reason it is called elbow method is because we see that around 6 to 8 we see an elbow occuring. That is why it is a good number to choose for our number of clusters. Again there isn't any correct k value, so we might have to rely on intuition and domain knowledge depending on the problem we are addressing with the help of unlabelled data in order to choose a correct or reasonable K value.
