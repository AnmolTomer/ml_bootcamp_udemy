## Neural Network Theory

- Neural networks are modeled after biological neural networks and attempt to allow computers to learn in a similar manner to humans through reinforcement learning.

- Use Cases for Neural Networks :
    - Pattern Recognition
    - Time Series Predictions
    - Signal Processing
    - Anomaly Detection
    - Self-Driving Vehicles

### Why Neural Networks ?

- There are problems which are difficult or too tedious for humans but really easy for computers, if given a specific set of tasks and algorithm of approaching the problems including the problems which are iterative in nature. e.g. Calculating large arithmetic numbers.

- Other way around, there are problems easy for humans, but difficult for computers. e.g. Recognizing a picture of a person from the side.

- Neural Networks attempt to solve problems that would normally be easy for humans but hard for computers!

### Comparison With Human Brain



- Human Brain has interconnected neurons with dendrites that receive inputs, and then based on those inputs, produce an electrical signal output through the axon to another neuron.
- We try to create the above idea through artificial Neural Network. (ANN)

## Simples Neural Network - The Perceptron

- A perceptron consists of one or more inputs, a processor and a single output.
- A perceptron follows the "feed forward" model, meaning inputs are sent into the neuron, are processed, and results in an output.


### Perceptron Process

- A perceptron process follows 4 main steps : 
    1. Receive Inputs
    2. Weigh Inputs
    3. Sum Inputs
    4. Generate Output

### Example :
Say we have a perceptron with 2 inputs as depicted above

Input 0: x1 = 12
Input 1: x2 = 4

Now, each input that is sent to the neuron first must be weighted i.e. multiplied by some value (often a number between -1 and 1)

When creating a perceptron, we'll typically begin by assigning random weights

**Weight 0 : 0.5**

**Weight 1 : -1**

We take each input and multiply by its weight. 

**Input 0 * Weight 0 => 12*0.5 = 6**

**Input 1 * Weight 1 => 4*-1 = -4**

Now, the output of a perceptron is generated by passing that sum through an activation function. In the case of a simple binary output, the activation function is what tells the perceptron whether to "fire" or not.

There are many activation functions to choose from (Logistic, Trigonometric, Step etc.). Let's make the activation function the sign of the sum. In other words if the sum is a positive number, the output is 1, if negative then the output is -1.



### What is Bias and its need ?

Another thing to consider is Bias. Imagine the both inputs were equal to zero, then any sum no matter what multiplicative weight would also be zero!

To avoid this problem, we add a third input known as a bias input with a value of 1. This avoids the zero issue.


## Steps to train the perceptron :
1. First we provide the perceptron with inputs for which there is a known answer.
2. Ask the perceptron to guess an answer.
3. Compute the error. (How far off from the correct answer predicted answer was.)
4. Adjust all the weights according to the error.
5. Return to step 1 and repeat.

Layering of a bunch of perceptrons is what ultimately leads to creation of a neural network.

- We keep on repeating this until we reach an error rate we are satisfied with (this we set beforehand).

That is how a single perceptron would work, now to create a neural network all we have to do is link many perceptrons together in layers.

## Layers in Neural Networks

- Generally, we have an input layer and an output layer. Any layers in between are known as hidden layers, because you don't directly see anything but the input or output.


- You might have heard of the term "Deep Learning". That's just a neural network with many hidden layers, causing it to be "deep". For example, Microsoft's state of the art vision recognition uses 152 layers.
