## What is SVM ?

- Support Vector Machines (SVMs) are supervised learning models with associated learning algorithms that analyze data and recognize patterns used for classification and regression analysis.
- We will be focusing on use of SVM for the purpose of classification.

- Given a set of training examples, each marked for belongging to one of two categories, a SVM training algorithm builds a model that assigns new examples into one category or the other, making it a non probabilistic, binary linear classifier.

- A SVM model is a representation of the examples as points in space, mapped so that the examples of the separate categories are divided by a clear gap that is as wide as possible.

- New examples are then mapped into that same space and predicted to belong to a category based on which side of the gap they fall on.

## Intuition behind SVMs

- Let's try to understand the basic intuition behind SVMs. Imagine the labelled training data below :



- We have 2 separate classes above, blue and pink. What we try to do is a binary classification for new points.
- We want to put a new point on this plot where horizontal axis is feature1 and y axis is feature2.
- We have to predict whether point belongs to blue class or pink class.
- We can draw a separate hyper-plane, in case of two dimensions, it is just a line, between 2 classes.


- Now we have many options of hyperplanes, that separate the two classes perfectly.




- **How to actually choose the line which separates the two classes perfectly ?**
- For a start, aim for choosing a hyperplane that maximizes the margin between classes.


- Dotted line is hyperplane, and there are margins extending from hyperplane.
- The vector points that the margin lines touch are known as Support Vectors.

- Support vectors are those training points which touch the margin line.

- We can expand this idea to non-linearly separable data through the "kernel trick."

- We have a X-label and Y-label in the 2D plot above, and we can notice that this data is not linearly separable, as it is almost a circle of inner blue and outer red part.
- So through kernel trick we end up visualising the data in a higher dimension.
