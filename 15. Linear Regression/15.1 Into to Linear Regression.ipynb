## History

* This all started in the 1800s with a guy named Francis Galton. Galton was studying the relationship between parents and their children. In particular, he investigated the relationship between the heights of fathers and their sons.

* What he discovered was that a man's son tended to be roughly as tall as his father. However Galton's breakthrough was that the son's height tended to be closer to the overall average height of all people.


## Example 1

* Let's take Shaquille O'Neal as an example. Shaq is really tall:7ft 1in (2.2 meters). If Shaq has a son, chances are he'll be pretty tall too. However, Shaq is such an anomaly that there is also a very good chance that his son will be not be as tall as Shaq.


* Turns out this is the case: Shaq's son is pretty tall (6 ft 7 in), but not nearly as tall as his dad. Galton called this phenomenon regression, as in **"A father's son's height tends to regress (or drift towards) the mean (average) height."**



## Example 2

* Let's take the simplest possible example : calculation a regression with only 2 data points Here we have 2 data points (2,4) and (5,10). 


* What we are trying to do when we calculate the regression line is to draw a line which is as close to every dot as possible. For classic linear regression, or "Least Square Method" you only measure the closeness in the "up and down" direction. Below we can see a perfectly fit line because we have only 2 data points.



## Example 3

* Now in a real world scenario we would like to apply the same to more than just 2 data points. By doing this we could take heights of multiple men and their son's and tell a man how tall we expect his son to be...even before he has a son.



* This is the idea behind supervised learning, given a bunch of labelled data points, we create a model (in our case linear regression )and try to take unlabelled data (father's height) and expect a labelled data (son's height) predicted.

* Our goal with linear regression is to minimize the vertical distance between all the data points and our line.

* In search of determining the best line of fit, we are attempting to minimize the distance between all the points and their distance to our line.

* There are different ways to minimize this sum, sum of squared errors, sum of absolute errors, but all these methods have a general goal of minimizing this distance, between line we have and resst of the data points.



## Example 4

* One of the most popular methods is the least squares method. Say we have some blue points on x-y 2D plane.


* Now we want to fit a linear regression line, how to decide which line is the one with best fit ?


* We will use the least squares method, which is fitted by minimizing the sum of squares of the residuals. The residual for an observation is the difference between the observation (y-value) and the fitted line.
